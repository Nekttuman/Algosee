```python
import matplotlib.pyplot as plt

x = [2, 10, 20, 40, 60, 80, 100, 1000, 10000, 20000, 30000]#, 40000, 50000, 70000, 100000]  # data len
```

```python
paths = {"bubble": "./tests_time/bubble_sort.txt", "shell":"./tests_time/shell_sort.txt", 
        "insertion":"./tests_time/insertion_sort.txt", "merge":"./tests_time/merge_sort.txt", 
        "quick":"./tests_time/quick_sort.txt", "selection":"./tests_time/selection_sort.txt",
        "STL sort":"./tests_time/std_sort.txt"}

values = {"bubble": {"random":[], "asc":[], "desc":[]}, "shell":{"random":[], "asc":[], "desc":[]}, 
        "insertion":{"random":[], "asc":[], "desc":[]}, "merge":{"random":[], "asc":[], "desc":[]}, 
        "quick":{"random":[], "asc":[], "desc":[]}, "selection":{"random":[], "asc":[], "desc":[]},
        "STL sort":{"random":[], "asc":[], "desc":[]}}

for key in paths:
    f = open(paths[key])
    l = [i for i in f.readline().split()]
    for s in l:
        k = [int(i) for i in s.split(':')]
        values[key]["random"].append(k[0])
        values[key]["asc"].append(k[1])
        values[key]["desc"].append(k[2])
```

```python
for key in paths:
    plt.plot(x[:], values[key]["random"], label=key)
plt.legend()
plt.savefig("./graphics/pic1.png")
```

```python
paths = {"merge":"./tests_time/merge_sort.txt", "STL sort":"./tests_time/std_sort.txt"}

for key in paths:
    plt.plot(x[:],values[key]["random"], label=key)
plt.legend()
plt.savefig("./graphics/pic3.png")
```

```python
sortType = "bubble"

plt.plot(x[:],values[sortType]["random"], label="random")
plt.plot(x[:],values[sortType]["asc"], label="best")
plt.plot(x[:],values[sortType]["desc"], label="worst")
plt.legend()

plt.title("Bubble sort")
plt.savefig("./graphics/"+sortType+".png")
```
