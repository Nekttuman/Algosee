```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('Solarize_Light2')


timeData = [22106564, 4535171, 83632069, 154782848, 97110487,481111354, 335063176, 179718984]
sizeData = [510896475, 8, 186289003, 736934975, 8, 1581475260, 8, 24372304]
```

```python
x = ['AH','AC', 'LZ78', 'MTF-HA', 'MTF-AC', 'rle-mtf-rle-ha', 'rle-mtf-rle-ac', 'PPM']

fig, ax = plt.subplots()

ax.bar(x, timeData)


fig.set_figwidth(12)    #  ширина Figure
plt.ylabel("Times mcs")

plt.savefig("time.png")
plt.show()
```

```python
fig, ax = plt.subplots()

ax.bar(x, sizeData, color="hotpink")


fig.set_figwidth(12)    #  ширина Figure
# fig.set_figheight(6)    #  высота Figure

plt.ylabel("data size Mb")
plt.savefig("size.png")
plt.show()
```
